#IMPORT LIBRIARIES AND DATASET

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import mglearn
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = make_blobs(random_state=42)
```

#VISUALIZATION WITH MGLEARN AND MATPLOTLIB

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1", "Class 2"])
```

#CREATE SVC MODEL AND FIR HER

```python
modelSVC = LinearSVC()
modelSVC.fit(X, y)
```

#FINAL VISUALIZATION

```python
mglearn.discrete_scatter(X[:,0], X[:,1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(modelSVC.coef_,
                                  modelSVC.intercept_, ['b', 'r', 'g']):
  plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.ylim(-10, 15)
plt.xlim(-10, 8)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(['Class 0', 'Class 1', 'Class 2', 'Class line 0', 'Class line 1', 'Class line 2'], loc = (1.01, 0.3))
```

```python
mglearn.plots.plot_2d_classification(modelSVC, X, fill=True, alpha=0.7)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
line = np.linspace(-15, 15)
for coef, intercept, color in zip(modelSVC.coef_, modelSVC.intercept_,
                                  ['b', 'r', 'g']):
  plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)
plt.legend(['Class 0', 'Class 1', 'Class 2',
            'Class line 0', 'Class line 1', 'Class line 2'], loc=(1.01, 0.3))
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
```
